```python
df_perc = df_survey.copy(deep=True)
df_perc['Very interested'] = df_perc['Very interested'].apply(lambda x: round(x*(1/2233),4)*100)
df_perc['Somewhat interested'] = df_perc['Somewhat interested'].apply(lambda x: round(x*(1/2233),4)*100)
df_perc['Not interested'] = df_perc['Not interested'].apply(lambda x: round(x*(1/2233),4)*100)
df_perc
```

```python
ax = df_perc.plot(kind = 'bar', figsize=(20,8),color = ['#5cb85c', '#5bc0de','#d9534f'], width = 0.8)
ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize = 16)
ax.legend(fontsize = 14)
plt.xticks(fontsize = 14)
#plt.yticks(fontsize = 14)
ax.set_yticklabels([])
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.set_ylabel('% of Respondents', fontsize =14)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    height = height.round(2)
    heightperc = str(height)+'%'
    x, y = p.get_xy()
    ax.annotate( (heightperc),(p.get_x()+.15*width, p.get_y() + height + 2.25), fontsize=14)
plt.show()
```
